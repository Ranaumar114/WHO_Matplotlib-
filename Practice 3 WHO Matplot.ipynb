```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
WHO = pd.read_csv('WHO.csv' , encoding = 'UTF-8')
```

```python
WHO
```

```python
country = WHO['Country'].unique()
country
```

```python
count_cumu = pd.Series()

for i in country:
    df = WHO[WHO['Country'] == i]
    df.sort_cumu = df.sort_values('Cumulative_cases' , ascending = False)
    cumulative = df.sort_cumu.iloc[0,3]
    count_cumu[i] = cumulative
    
count_cumu
```

```python
WHO['Date_reported'] = pd.to_datetime(WHO['Date_reported'])
WHO.info()
```

```python
df = WHO[WHO['Country'] == 'Pakistan']
df
```

```python
df.sort_cumu = df.sort_values('Cumulative_cases' , ascending = False)
df.sort_cumu
```

```python
plt.plot(df['Date_reported'] , df['Cumulative_cases'])

plt.title('Pakistan')
plt.xlabel('Date')
plt.ylabel('Cumulative Cases')

plt.show()
```

```python
country_name = 'India'
f'{country_name} : Cumulative Reports'
```

```python
def plot_country(country_name):
    
    df = WHO[WHO['Country'] == country_name]
    
    plt.plot(df['Date_reported'] , df['Cumulative_cases'])
    plt.title(f'{country_name} : Cumulative Cases')
    plt.ticklabel_format(axis = 'y' , style = 'plain')
    plt.xlabel('Date')
    plt.ylabel('Number of Cases')
    
    plt.show()
```

```python
plot_country('India')
```

```python
for i in WHO['Country'].unique()[:10]:
    plot_country(i)
```

```python
france = WHO[WHO['Country'] == 'France']
pak = WHO[WHO['Country'] == 'Pakistan']
ind = WHO[WHO['Country'] == 'India']
```

```python
plt.plot(france['Date_reported'] , france['Cumulative_cases'] , color = 'yellow' , label = 'FRANCE')
plt.plot(pak['Date_reported'] , pak['Cumulative_cases'] , color = 'green' , label = 'PAKISTAN')
plt.plot(ind['Date_reported'] , ind['Cumulative_cases'], color = 'red' , label = 'INDIA')

plt.title('Comparison')
plt.ticklabel_format(axis = 'y' , style = 'plain')
plt.xlabel('Date')
plt.ylabel('No. of Cases')


plt.legend()
plt.show()
```
